### Blocks

We're going to be using this Python library for visualising arrays later. Execute the cells below and then edit them and see what you can do.

```python
from ipythonblocks import BlockGrid
```

```python
# Make a 5x5 grid
grid = BlockGrid(5,5)
# For each block in the grid
for block in grid:
    # Paint it red
    block.rgb = (255,0,0)
grid
```

```python
grid = BlockGrid(5,5)
for block in grid:
    if block.col == 3:
        block.rgb = (255,0,0)
    elif block.row < 2:
        block.rgb = (0,0,255)
    else:
        block.rgb = (0,255,0)
grid
```
